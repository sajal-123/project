```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df=pd.read_csv('shopping_trends_updated.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
print(f"The unique values in Category column,{df['Category'].unique()}")
print()
print(f"The unique values in Location column,{df['Location'].unique()}")
print()
print(f"The unique values in Color column,{df['Color'].unique()}")
print()
print(f"The unique values in Season column,{df['Season'].unique()}")
print()
print(f"The unique values in Shipping Type column,{df['Shipping Type'].unique()}")
print()
print(f"The unique values in Payment Method column,{df['Payment Method'].unique()}")
print()
print(f"The unique values in Frequency of Purchases column,{df['Frequency of Purchases'].unique()}")
```

```python
df['Age'].value_counts()
```

```python
df['Age Category']=pd.cut(df['Age'],bins=[0,15,18,30,50,70],labels=['child','teen','Young Adults','MiddleAge Adults','old'])
```

```python
categories = df['Age Category'].unique()
for category in categories:
    subset = df[df['Age Category'] == category]
    plt.hist(subset['Age'], bins=5, alpha=0.5, label=category)

plt.title('Age Distribution by Category')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(title='Age Category')
plt.show()
```

# Which Gender has the highest value of purchase

```python
import seaborn as sns
```

```python
df.groupby('Category')['Item Purchased'].value_counts()
```

```python
categories = df['Category'].unique()

for category in categories:
    subset = df[df['Category'] == category]
    plt.hist(subset['Item Purchased'], bins=5, alpha=0.5, label=f'Category {category}')

plt.title('Item Purchased Distribution by Category', fontsize=14, fontweight='bold')
plt.xlabel('Item Purchased', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10) 
plt.legend(title='Category', fontsize=10, title_fontsize=12) 
plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.tight_layout()
plt.show()
```

# compare do between customer spending time with or without promo codes

```python
df.groupby('Promo Code Used')['Purchase Amount (USD)'].mean().sort_values(ascending=False)
```

```python
# Grouped bar plot with Seaborn
sns.set(style="whitegrid")

plt.figure(figsize=(5, 3))
sns.barplot(
    data=df, 
    x="Promo Code Used", 
    y="Purchase Amount (USD)", 
    hue="Category", 
    palette="viridis"
)

# Beautify the plot
plt.title("Purchase Amount by Promo Code and Category", fontsize=14, fontweight="bold")
plt.xlabel("Promo Code Used", fontsize=12)
plt.ylabel("Total Purchase Amount (USD)", fontsize=12)
plt.legend(title="Category", fontsize=10)
plt.tight_layout()
plt.show()
```
